### Avg time between payments per account

```python
import pandas as pd
```

```python
payments = pd.read_csv('../Resources/payments.csv')
# payments.head()
```

```python
# Convert to datetime
payments['effective_when'] = pd.to_datetime(payments['effective_when'])

# Sort by date for each account
payments = payments.sort_values(['account_id','effective_when'])

# Reset index
payments = payments.reset_index(drop=True)
```

```python
# Calculate time in between payments
time_diff = [payments['effective_when'][i+1]-payments['effective_when'][0+i] for i in range(0,len(payments)-1)]

# Insert None for first row
time_diff.insert(0,None)

# Add calculated column
payments['time_between_payments'] = time_diff
```

```python
# Convert to seconds
payments['time_between_payments'] = payments['time_between_payments'].dt.total_seconds()

# Drop negative time diffs
payments = payments.loc[payments['time_between_payments']>0]
```

```python
# Group by id and find average mean of time between payments
payments_grouped = payments.groupby('account_id',as_index=False)['time_between_payments'].mean()

# Rename column
payments_grouped = payments_grouped.rename(columns={'time_between_payments':'avg_time_between_payments'})

# Convert back to days by dividing seconds by 86,400 (60*60*24 - minutes,hours,days)
payments_grouped['avg_time_between_payments'] = payments_grouped['avg_time_between_payments']/86400
```
